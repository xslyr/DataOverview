```python
import pandas as pd
import dfview as vw  # <--- our package
```

```python
df = pd.read_csv('data.csv')
```

#### Initializing our DataOverview objet linking to "my_describe" variable

```python
my_describe = vw.DataOverview(df)
```

#### Printing of column's type distributions

```python
my_describe.columns_type()
```

#### Printing our describe data for object columns

```python
my_describe.show('object')
```

#### Showing the data distribution of a column list 

```python
my_describe.obj_distrib(column_list=['MSZoning','Street','Alley','LotShape','LandContour'])
```

```python
my_describe.obj_distrib(column_list=['MSZoning','Street','Alley','LotShape','LandContour'], axis=0)
```

```python
my_describe.obj_distrib(column_list=['MSZoning','Street','Alley','LotShape','LandContour'], axis=0, include_nulls=False)
```

The axis with numbers is used to count how many occurence of each data type at that column
